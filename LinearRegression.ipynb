# Some Linear Data

```python
import random

random.seed(54)

X = [random.random() for i in range(1000)]
y = [4 + 3 * x + random.gauss(0, 1) for x in X]
```

# LinearRegresion with Gradient descent

```python
def LinearRegression(lr, n_iters, X, y):
        a = 0
        b = 0
        n = len(X)
        for i in range(n_iters):
           a_grad = 0
           b_grad = 0
           for m in range(n):
              xi = X[m]
              yi = y[m]
              a_grad += -(2/n) * xi * (yi - (a * xi + b))
              b_grad += -(2/n) * (yi - (a * xi + b))
           a = a - lr * a_grad
           b = b - lr * b_grad
        return a, b
```

```python
a,b = LinearRegression(0.1, 100, X, y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
x_line = [min(X), max(X)]
y_line = [a * x + b for x in x_line]
plt.plot(x_line, y_line, color='red')

plt.show()
```
